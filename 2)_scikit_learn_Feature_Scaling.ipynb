# Applying Normalization : Minmax Scaling

---


why we need to feature scale: so that all the columns have similar weightag/values of all the variables are to a similar scale

MinMax Scaling results in the range of 0 and 1. Also known as Normalization

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
```

```python
scald_df=scaler.fit_transform(df)
```

```python
scald_df
```

```python
scald_df=pd.DataFrame(data=scald_df,columns=df.columns)
```

```python
scald_df
```
