# Standardization: Z-score Normalization


scaling such that mean of features is 0 and standard deviation is 1

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/content/Social_Network_Ads.csv')
```

```python
df
```

```python
df.drop('User ID', axis=1, inplace=True)
```

```python
df['Gender'] = df['Gender'].map({'Male':1,'Female':0})
```

```python
df
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=df.drop('Purchased',axis=1)
```

```python
y=df['Purchased']
```

```python
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
scaler=StandardScaler()
```

```python
X_train_scald=scaler.fit_transform(X_train)
```

```python
X_test_scald=scaler.fit_transform(X_test)
```

```python
X_test_scladpd=pd.DataFrame(X_test_scald,columns=X_test.columns)
```

```python
X_train_scaldpd=pd.DataFrame(X_train_scald,columns=X_train.columns)
```
